```python
# Math library
import math
import numpy as np
# Plotting library
import matplotlib.pyplot as plt
```

```python
# Create some input / output data
x = np.array([0.03, 0.19, 0.34, 0.46, 0.78, 0.81, 1.08, 1.18, 1.39, 1.60, 1.65, 1.90])
y = np.array([0.67, 0.85, 1.05, 1.0, 1.40, 1.5, 1.3, 1.54, 1.55, 1.68, 1.73, 1.6 ])

print(x)
print(y)
```

```python
# Define 1D linear regression model
def f(x, phi0, phi1):
  # TODO :  Replace this line with the linear regression model (eq 2.4)
  #y = x
  y = phi0 + (phi1*x)

  return y
```

```python
# Function to help plot the data
def plot(x, y, phi0, phi1):
    fig,ax = plt.subplots()
    ax.scatter(x,y)
    plt.xlim([0,2.0])
    plt.ylim([0,2.0])
    ax.set_xlabel('Input, $x$')
    ax.set_ylabel('Output, $y$')
    # Draw line
    x_line = np.arange(0,2,0.01)
    print(x_line)
    y_line = f(x_line, phi0, phi1)
    plt.plot(x_line, y_line,'b-',lw=2)

    plt.show()
```

```python
# Set the intercept and slope as in figure 2.2b
phi0 = 0.4 ; phi1 = 0.2
# Plot the data and the model
plot(x,y,phi0,phi1)
```

```python
# Function to calculate the loss
def compute_loss(x,y,phi0,phi1):

  # TODO Replace this line with the loss calculation (equation 2.5)
  loss = 0
  for i in range(len(x)):
      # Predicted value using phi0 and phi1
    predicted = phi0 + phi1 * x[i]
       # Squared difference between predicted and actual values
    loss += (predicted - y[i]) ** 2


  return loss

    
```

```python
# Compute the loss for our current model
loss = compute_loss(x,y,phi0,phi1)
print(f'Your Loss = {loss:3.2f}, Ground truth =7.07')
```

```python
def finding_loss_automatically(x, y, phi0, phi1, learning_rate, iterations):
    phi0_history = [phi0]
    phi1_history = [phi1]
    gradient_phi0_history = []
    gradient_phi1_history = []
    loss_history = []

    for i in range(iterations):        
        current_loss = compute_loss(x, y, phi0, phi1)
        loss_history.append(current_loss)
        
        phi0_new = phi0 + learning_rate
        loss_phi0 = compute_loss(x, y, phi0_new, phi1)

        gradient_phi0 = (loss_phi0 - current_loss) / learning_rate

        phi1_new = phi1 + learning_rate
        loss_phi1 = compute_loss(x, y, phi0, phi1_new)

        gradient_phi1 = (loss_phi1 - current_loss) / learning_rate

        phi0 -= learning_rate * gradient_phi0
        phi1 -= learning_rate * gradient_phi1

        phi0_history.append(phi0)
        phi1_history.append(phi1)
        gradient_phi0_history.append(gradient_phi0)
        gradient_phi1_history.append(gradient_phi1)

    return phi0, phi1, current_loss, phi0_history, phi1_history, gradient_phi0_history, gradient_phi1_history, loss_history

phi0, phi1, current_loss, phi0_history, phi1_history, gradient_phi0_history, gradient_phi1_history, loss_history = finding_loss_automatically(x, y, phi0, phi1, 0.01, 1000)
plot(x,y,phi0,phi1)
print(f'Your Loss = {current_loss:3.2f}')
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

# Function to plot gradients for phi0
def plot_3d_scatter_gradient_phi0(phi0_history, phi1_history, gradient_phi0_history):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Convert histories to arrays for plotting
    phi0_history = np.array(phi0_history)
    phi1_history = np.array(phi1_history)
    gradient_phi0_history = np.array(gradient_phi0_history)

    # Create a scatter plot in 3D for gradient of phi0
    scatter = ax.scatter(phi0_history[:-1], phi1_history[:-1], gradient_phi0_history, c=gradient_phi0_history, cmap='plasma', s=50, alpha=0.8)

    # Connect points with a line to show the trajectory
    ax.plot(phi0_history[:-1], phi1_history[:-1], gradient_phi0_history, color='black', lw=1, alpha=0.5)

    # Add color bar
    fig.colorbar(scatter, ax=ax, label='Gradient of phi0')

    # Set axis labels
    ax.set_xlabel('phi0')
    ax.set_ylabel('phi1')
    ax.set_zlabel('Gradient of phi0')

    ax.grid(True)
    plt.title('3D Scatter Plot of Gradient of phi0')
    plt.show()

# Function to plot gradients for phi1
def plot_3d_scatter_gradient_phi1(phi0_history, phi1_history, gradient_phi1_history):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Convert histories to arrays for plotting
    phi0_history = np.array(phi0_history)
    phi1_history = np.array(phi1_history)
    gradient_phi1_history = np.array(gradient_phi1_history)

    # Create a scatter plot in 3D for gradient of phi1
    scatter = ax.scatter(phi0_history[:-1], phi1_history[:-1], gradient_phi1_history, c=gradient_phi1_history, cmap='viridis', s=50, alpha=0.8)

    # Connect points with a line to show the trajectory
    ax.plot(phi0_history[:-1], phi1_history[:-1], gradient_phi1_history, color='black', lw=1, alpha=0.5)

    # Add color bar
    fig.colorbar(scatter, ax=ax, label='Gradient of phi1')

    # Set axis labels
    ax.set_xlabel('phi0')
    ax.set_ylabel('phi1')
    ax.set_zlabel('Gradient of phi1')

    ax.grid(True)
    plt.title('3D Scatter Plot of Gradient of phi1')
    plt.show()

# Plot for phi0 gradient
plot_3d_scatter_gradient_phi0(phi0_history, phi1_history, gradient_phi0_history)

# Plot for phi1 gradient
plot_3d_scatter_gradient_phi1(phi0_history, phi1_history, gradient_phi1_history)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

# Compute loss over a range of phi0 and phi1 values
def compute_loss_surface(x, y, phi0_range, phi1_range):
    phi0_mesh, phi1_mesh = np.meshgrid(phi0_range, phi1_range)
    loss_mesh = np.zeros_like(phi0_mesh)
    
    # Calculate loss for each pair of phi0 and phi1 in the meshgrid
    for i in range(len(phi0_range)):
        for j in range(len(phi1_range)):
            loss_mesh[i, j] = compute_loss(x, y, phi0_mesh[i, j], phi1_mesh[i, j])
    
    return phi0_mesh, phi1_mesh, loss_mesh

# Plot the gradient for phi0
def plot_gradient_phi0(x, y, phi0_history, phi1_history, gradient_phi0_history):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Define ranges for phi0 and phi1 to compute the loss surface
    phi0_range = np.linspace(min(phi0_history) - 1, max(phi0_history) + 1, 100)
    phi1_range = np.linspace(min(phi1_history) - 1, max(phi1_history) + 1, 100)

    # Compute the loss surface
    phi0_mesh, phi1_mesh, loss_mesh = compute_loss_surface(x, y, phi0_range, phi1_range)

    # Plot the loss surface (mesh)
    ax.plot_surface(phi0_mesh, phi1_mesh, loss_mesh, cmap='viridis', alpha=0.7)

    # Correct the length mismatch
    phi0_history = np.array(phi0_history[:-1])
    phi1_history = np.array(phi1_history[:-1])
    gradient_phi0_history = np.array(gradient_phi0_history)

    # Plot the gradient descent path for phi0
    scatter = ax.scatter(phi0_history, phi1_history, gradient_phi0_history, c=gradient_phi0_history, cmap='plasma', s=50)

    # Add arrows to represent the gradient direction for phi0
    for i in range(1, len(phi0_history)):
        ax.quiver(phi0_history[i-1], phi1_history[i-1], gradient_phi0_history[i-1],
                  phi0_history[i] - phi0_history[i-1], phi1_history[i] - phi1_history[i-1], 
                  gradient_phi0_history[i] - gradient_phi0_history[i-1],
                  color='black', alpha=0.8, arrow_length_ratio=0.2)

    # Add color bar and labels
    fig.colorbar(scatter, ax=ax, label='Gradient of phi0')
    ax.set_xlabel('phi0')
    ax.set_ylabel('phi1')
    ax.set_zlabel('Gradient of phi0')

    plt.title('Gradient Descent Path for phi0 on Loss Surface')
    plt.show()

# Plot the gradient for phi1
def plot_gradient_phi1(x, y, phi0_history, phi1_history, gradient_phi1_history):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Define ranges for phi0 and phi1 to compute the loss surface
    phi0_range = np.linspace(min(phi0_history) - 1, max(phi0_history) + 1, 100)
    phi1_range = np.linspace(min(phi1_history) - 1, max(phi1_history) + 1, 100)

    # Compute the loss surface
    phi0_mesh, phi1_mesh, loss_mesh = compute_loss_surface(x, y, phi0_range, phi1_range)

    # Plot the loss surface (mesh)
    ax.plot_surface(phi0_mesh, phi1_mesh, loss_mesh, cmap='viridis', alpha=0.7)

    # Correct the length mismatch
    phi0_history = np.array(phi0_history[:-1])
    phi1_history = np.array(phi1_history[:-1])
    gradient_phi1_history = np.array(gradient_phi1_history)

    # Plot the gradient descent path for phi1
    scatter = ax.scatter(phi0_history, phi1_history, gradient_phi1_history, c=gradient_phi1_history, cmap='plasma', s=50)

    # Add arrows to represent the gradient direction for phi1
    for i in range(1, len(phi0_history)):
        ax.quiver(phi0_history[i-1], phi1_history[i-1], gradient_phi1_history[i-1],
                  phi0_history[i] - phi0_history[i-1], phi1_history[i] - phi1_history[i-1], 
                  gradient_phi1_history[i] - gradient_phi1_history[i-1],
                  color='black', alpha=0.8, arrow_length_ratio=0.2)

    # Add color bar and labels
    fig.colorbar(scatter, ax=ax, label='Gradient of phi1')
    ax.set_xlabel('phi0')
    ax.set_ylabel('phi1')
    ax.set_zlabel('Gradient of phi1')

    plt.title('Gradient Descent Path for phi1 on Loss Surface')
    plt.show()

# Plot gradient for phi0
plot_gradient_phi0(x, y, phi0_history, phi1_history, gradient_phi0_history)

# Plot gradient for phi1
plot_gradient_phi1(x, y, phi0_history, phi1_history, gradient_phi1_history)
```

# Visualizing the loss function

The above process is equivalent to descending coordinate wise on the loss function<br>

```python
# Make a 2D grid of possible phi0 and phi1 values
phi0_mesh, phi1_mesh = np.meshgrid(np.arange(0.0,2.0,0.02), np.arange(-1.0,1.0,0.02))

# Make a 2D array for the losses
all_losses = np.zeros_like(phi1_mesh)
# Run through each 2D combination of phi0, phi1 and compute loss
for indices,temp in np.ndenumerate(phi1_mesh):
    all_losses[indices] = compute_loss(x,y, phi0_mesh[indices], phi1_mesh[indices])
```

```python
# Plot the loss function as a heatmap
fig = plt.figure()
ax = plt.axes()
fig.set_size_inches(7,7)
levels = 256
ax.contourf(phi0_mesh, phi1_mesh, all_losses ,levels)
levels = 40
ax.contour(phi0_mesh, phi1_mesh, all_losses ,levels, colors=['#80808080'])
ax.set_ylim([1,-1])
ax.set_xlabel(r'Intercept, $\phi_0$')
ax.set_ylabel(r'Slope, $\phi_1$')

# Plot the position of your best fitting line on the loss function
# It should be close to the minimum
ax.plot(phi0,phi1,'ro')
plt.show()
```
